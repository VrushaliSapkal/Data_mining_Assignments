# Visualization Technique1- Histogram 

```python
data.hist()
plt.show()
```

# 2 : Density Plot

```python
#Density Plots
#A density plot appears to be an abstracted histogram. Each bin has a smooth curve drawn through its top
data.plot(kind='density',subplots=True,sharex=False)
plt.show()
```

# 3 : Box Plot

```python
#A box plot summarizes how each attribute is distributed.
data.plot(kind='box',subplots=True,sharex=False,sharey=False)
```

# 4 : ScatterMatrix

```python
pd.plotting.scatter_matrix(data)
```

# 5 : Bar Plot

```python
import seaborn as sns
```

```python
sns.barplot(data['Volume'],data['Open'])
```

```python
# Scatter plot with day against tip
plt.scatter(data['High'], data['Low'], c=data['Open'],
            s=data['Close'])
 
# Adding Title to the Plot
plt.title("Scatter Plot")
 
# Setting the X and Y labels
plt.xlabel('High')
plt.ylabel('Low')
 
plt.colorbar()
 
plt.show()
```
